```python
import z3
f = open('../benchmarks/quantification/blazer_k96_unsafe/combined_cache_cbmc.dimacs', 'r')
target_var = "label_alignment_$3"
first_line = f.readline()
splits = first_line.split(" ")
literal_num = int(splits[2])
clause_num = int(splits[3])

clauses = f.readlines()

comments = [c for c in clauses if c[0] == 'c']
comments_on_public = [c for c in comments if "_pub!" in c]
pub_literals = [c.split(" ")[2:] for c in comments_on_public]
import itertools
pub_literals = list(itertools.chain.from_iterable(pub_literals))
pub_literals = [l.strip() for l in pub_literals]
target_literals = [c.split(" ")[2:] for c in comments if target_var in c]
target_literals = list(itertools.chain.from_iterable(target_literals))
target_literals = [l.strip() for l in target_literals]


clauses = [c for c in clauses if c[0] != 'c']
print(len(pub_literals), len(target_literals))
```

```python
comments_on_public
pub_literals
```

```python
dup_clauses = []
for c in clauses:
    splits = c.split(" ")
    dup_clauses.append(" ".join([str(int(l) + literal_num) if int(l) > 0 else str(int(l) - literal_num) for l in splits[:-1]]) + " 0\n")
clauses += dup_clauses
s = z3.Solver()
s.from_string("p cnf {} {}\n".format(literal_num*2, clause_num*2) + "\n".join(clauses))
```

```python
from z3_enumerate import get_vars

vars_mapping = {}

for assertion in s.assertions():
    vs = get_vars(assertion)
    vars_mapping.update({str(v).strip("k!"): v for v in vs})

```

```python
for l in pub_literals:
    if l in vars_mapping and str(int(l) + literal_num) in vars_mapping:
        s.add(vars_mapping[l] == vars_mapping[str(int(l) + literal_num)])

for l in target_literals:
    if l in vars_mapping and str(int(l) + literal_num) in vars_mapping:
        s.add(vars_mapping[l] != vars_mapping[str(int(l) + literal_num)])

s.check()
```

```python
m = s.model()
for l in pub_literals:
    dup_l = str(int(l) + literal_num)
    if l in vars_mapping and dup_l in vars_mapping:

        print(m[vars_mapping[l]] == m[vars_mapping[dup_l]])


# str(m[vars_mapping[str(4+literal_num)]])
# m[vars_mapping[str(int(target_literals[0]) + literal_num)]]
```
